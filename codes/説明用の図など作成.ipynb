```python
#　漫画画像のメタデータから矩形領域を描画する．
import cv2
import os
import xml.etree.ElementTree as ET

def draw_rectangle(image, xmin, ymin, xmax, ymax, color=(0, 255, 0), thickness=2):
    cv2.rectangle(image, (xmin, ymin), (xmax, ymax), color, thickness)

def main():
    xml_file_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations.v2020.12.18/GOOD_KISS_Ver2.xml"
    images_folder_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/images/GOOD_KISS_Ver2/"
    output_folder_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/images_highlighted/face_text_frame_highlighted/"

    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    for page in root.iter("page"):
        page_index = int(page.get("index"))
        image_path = os.path.join(images_folder_path, f"{page_index:03d}.jpg")
        image = cv2.imread(image_path)

        for element in page:
            if element.tag in ["face", "text", "frame"]:
                xmin = int(element.get("xmin"))
                ymin = int(element.get("ymin"))
                xmax = int(element.get("xmax"))
                ymax = int(element.get("ymax"))

                if element.tag == "frame":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(255, 0, 0))  # Blue for frames
                elif element.tag == "text":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(0, 255, 0))  # Green for frames
                elif element.tag == "face":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(0, 0, 255))  # Red for frames
                else:
                    draw_rectangle(image, xmin, ymin, xmax, ymax)  # Default green for other elements

        output_path = os.path.join(output_folder_path, f"{page_index:03d}_highlighted.jpg")
        cv2.imwrite(output_path, image)

if __name__ == "__main__":
    main()

```

```python
#　漫画画像のメタデータから矩形領域を描画する．
import cv2
import os
import xml.etree.ElementTree as ET

def draw_rectangle(image, xmin, ymin, xmax, ymax, color=(0, 255, 0), thickness=2):
    cv2.rectangle(image, (xmin, ymin), (xmax, ymax), color, thickness)

def main():
    xml_file_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations.v2020.12.18/MukoukizuNoChonbo.xml"
    images_folder_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/images/MukoukizuNoChonbo/"
    output_folder_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/images_highlighted/MukoukizuNoChonbo_all_highlighted/"

    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    for page in root.iter("page"):
        page_index = int(page.get("index"))
        image_path = os.path.join(images_folder_path, f"{page_index:03d}.jpg")
        image = cv2.imread(image_path)

        for element in page:
            if element.tag in ["face", "text", "frame", "body"]:
                xmin = int(element.get("xmin"))
                ymin = int(element.get("ymin"))
                xmax = int(element.get("xmax"))
                ymax = int(element.get("ymax"))

                if element.tag == "frame":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(255, 0, 0))  # Blue for frames
                elif element.tag == "text":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(0, 255, 0))  # Green for frames
                elif element.tag == "face":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(0, 0, 255))  # Red for frames
                elif element.tag == "body":
                    draw_rectangle(image, xmin, ymin, xmax, ymax, color=(168, 87, 167))  # Purple for frames
                else:
                    draw_rectangle(image, xmin, ymin, xmax, ymax)  # Default green for other elements

        output_path = os.path.join(output_folder_path, f"{page_index:03d}_highlighted.jpg")
        cv2.imwrite(output_path, image)

if __name__ == "__main__":
    main()

```
