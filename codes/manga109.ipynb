```python
import xml.etree.ElementTree as ET

# XMLファイルのパス
xml_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations/ARMS.xml"

# XMLファイルを解析
tree = ET.parse(xml_path)
root = tree.getroot()

# 各ページごとにフレーム情報を格納するリスト
page_frames = []

# 各ページをループ
for page in root.findall('pages/page'):
    frames = []  # 各ページのフレーム情報を格納するリスト

    # ページ内のフレーム要素をループ
    for frame in page.findall('.//frame'):
        xmin = frame.get("xmin")
        ymin = frame.get("ymin")
        xmax = frame.get("xmax")
        ymax = frame.get("ymax")

        frame_info = {
            "xmin": xmin,
            "ymin": ymin,
            "xmax": xmax,
            "ymax": ymax
        }

        frames.append(frame_info)

    # ページごとのフレーム情報をリストに追加
    page_frames.append(frames)

# ページごとのフレーム情報を表示
for page_num, frames in enumerate(page_frames, start=1):
    print(f"ページ {page_num} のフレーム情報:")
    for idx, frame_info in enumerate(frames, start=1):
        print(f"  フレーム {idx}: xmin={frame_info['xmin']}, ymin={frame_info['ymin']}, xmax={frame_info['xmax']}, ymax={frame_info['ymax']}")
```

```python
from PIL import Image, ImageDraw
import xml.etree.ElementTree as ET
import os

# XMLファイルのパス
xml_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations/ARMS.xml"
# 画像ファイルが存在するディレクトリ
image_dir = "/content/drive/MyDrive/Manga109_released_2021_12_30/images/ARMS"  # 画像ファイルが存在するディレクトリのパスに置き換えてください

# XMLファイルを解析
tree = ET.parse(xml_path)
root = tree.getroot()

# 各ページごとにコマの矩形領域を画像に描画
for page_num, page in enumerate(root.findall('pages/page'), start=0):
    image_path = f"{image_dir}/{page_num:03d}.jpg"  # 画像ファイルのパスを生成

    # 画像ファイルが存在するか確認
    if os.path.exists(image_path):
        image = Image.open(image_path)
        draw = ImageDraw.Draw(image)

        # ページ内のフレーム要素をループ
        for frame in page.findall('.//frame'):
            xmin = int(frame.get("xmin"))
            ymin = int(frame.get("ymin"))
            xmax = int(frame.get("xmax"))
            ymax = int(frame.get("ymax"))

            # コマの矩形領域を表す線を描画
            draw.rectangle([xmin, ymin, xmax, ymax], outline="red", width=5)

        # 描画された画像を保存（オプション）
        output_image_path = f"{image_dir}/annotated_{page_num:03d}.jpg"
        image.save(output_image_path)

        # メモリを解放
        del draw
        del image
    else:
        print(f"画像 {image_path} が存在しません。")

# 描画が完了したらメモリを解放
del tree
del root
```

```python
import xml.etree.ElementTree as ET

# 新しいメタデータ
new_metadata = {
    'Subjective_Joy': '0',
    'Subjective_Sadness': '0',
    'Subjective_Anticipation': '0',
    'Subjective_Surprise': '0',
    'Subjective_Anger': '0',
    'Subjective_Fear': '0',
    'Subjective_Disgust': '0',
    'Subjective_Trust': '0',
    'Objective_Joy': '0',
    'Objective_Sadness': '0',
    'Objective_Anticipation': '0',
    'Objective_Surprise': '0',
    'Objective_Anger': '0',
    'Objective_Fear': '0',
    'Objective_Disgust': '0',
    'Objective_Trust': '0',
}

# 新しいメタデータを追加する関数
def add_metadata_to_frame(frame_elem):
    for key, value in new_metadata.items():
        frame_elem.set(key, value)

# 新しいXMLのルート要素を作成
new_root = ET.Element("frames")

# 元のXMLファイルを読み込み
xml_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations.v2020.12.18/Akuhamu.xml"
tree = ET.parse(xml_path)
root = tree.getroot()

# 各ページのフレーム情報を抽出
for page_elem in root.findall(".//page"):
    page_index = page_elem.get("index")
    page_width = page_elem.get("width")

    for frame_elem in page_elem.findall(".//frame"):
        frame_id = frame_elem.get("id")
        xmin = frame_elem.get("xmin")
        ymin = frame_elem.get("ymin")
        xmax = frame_elem.get("xmax")
        ymax = frame_elem.get("ymax")

        # 新しいXMLにフレーム情報を追加
        frame_info = ET.SubElement(new_root, "frame")
        frame_info.set("id", frame_id)
        frame_info.set("page_index", page_index)
        frame_info.set("page_width", page_width)
        frame_info.set("xmin", xmin)
        frame_info.set("ymin", ymin)
        frame_info.set("xmax", xmax)
        frame_info.set("ymax", ymax)

        # メタデータを追加
        add_metadata_to_frame(frame_info)

        # 改行を追加
        ET.SubElement(new_root, "").text = "\n"

# 新しいXMLを保存
new_tree = ET.ElementTree(new_root)
new_tree.write("/content/drive/MyDrive/Manga109_released_2021_12_30/annotations.v2020.12.18/Akuhamu_onlyframes.xml", encoding="utf-8", xml_declaration=True)

```

```python
import xml.etree.ElementTree as ET

# XMLファイルのパス
xml_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations/ARMS.xml"

# XMLデータの読み込み
tree = ET.parse(xml_path)
root = tree.getroot()

# ページごとにframeを取得し、ソートする関数
def sort_frames(page):
    frames = page.findall('.//frame')
    frames.sort(key=lambda frame: (int(frame.get('xmax')), int(frame.get('ymin'))))
    return frames

# ページごとにソートして出力
for page in root.findall('.//page'):
    page_index = page.get('index')
    frames = sort_frames(page)

    print(f'Page {page_index}:')
    for frame in frames:
        print(f'  Frame {frame.get("id")}: ({frame.get("xmax")}, {frame.get("ymin")})')

    page_width = int(page.get('width'))
    print(f'  Page Size: ({page_width}, 0)\n')

```

```python
def calculate_reading_time(text):
    # 文章の文字数をカウント
    char_count = len(text)

    # 1文字あたりの読むのにかかる時間（秒）
    time_per_char = 0.25

    # 総所要時間（秒）
    total_time = char_count * time_per_char

    return char_count, total_time

# テキストを入力
input_text = input("テキストを入力してください: ")

# 文字数と所要時間を計算
char_count, total_time = calculate_reading_time(input_text)

# 結果を出力
print(f"文字数: {char_count}")
print(f"所要時間（秒）: {total_time}")
```
