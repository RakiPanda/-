```python
import xml.etree.ElementTree as ET
import cv2
import os

# XMLファイルのパス
folder_path = "/content/drive/MyDrive/Manga109_released_2021_12_30/annotations.v2020.12.18/DualJustice.xml"

# フォルダ内のXMLファイルの数だけループ
for xml_filename in os.listdir(folder_path):
    if xml_filename.endswith(".xml"):
        xml_path = os.path.join(folder_path, xml_filename)

        try:
            # XMLファイルを読み込む
            tree = ET.parse(xml_path)
            root = tree.getroot()

            # XMLファイルの名前を取得
            xml_filename = os.path.basename(xml_path)

            # 拡張子を除いた部分を取得
            xml_name_without_extension = os.path.splitext(xml_filename)[0]
            print(xml_name_without_extension)

            # 画像が含まれているフォルダのパス 最後の/を忘れずに！
            image_folder = f"/content/drive/MyDrive/Manga109_released_2021_12_30/images/{xml_name_without_extension}/"

            # 保存先のフォルダのパス（例: /content/drive/MyDrive/SavedFaces/） 最後の/を忘れずに！
            save_folder = f"/content/drive/MyDrive/Manga109_released_2021_12_30/face_images/{xml_name_without_extension}/"

            # フォルダが存在しない場合は生成する
            os.makedirs(image_folder, exist_ok=True)
            os.makedirs(save_folder, exist_ok=True)

            # 各ページごとに処理
            for page_elem in root.findall(".//page"):
                page_index = page_elem.get("index").zfill(3)  # ページ番号を3桁の0埋め形式に変換
                image_filename = f"{page_index}.jpg"  # 対応する画像ファイル名

                # 画像を読み込む
                image_path = f"{image_folder}{image_filename}"
                img = cv2.imread(image_path)

                # 各face要素に対応する画像領域を切り抜いて保存
                for face_elem in page_elem.findall(".//face"):
                    face_id = face_elem.get("id").zfill(8)  # face IDを8桁の0埋め形式に変換
                    xmin = int(face_elem.get("xmin"))
                    ymin = int(face_elem.get("ymin"))
                    xmax = int(face_elem.get("xmax"))
                    ymax = int(face_elem.get("ymax"))

                    # 画像を切り抜いて保存
                    face_image = img[ymin:ymax, xmin:xmax]
                    face_filename = f"{face_id}.jpg"
                    face_path = f"{save_folder}{face_filename}"  # 保存先のパスを指定
                    cv2.imwrite(face_path, face_image)

                    # print(f"Saved: {face_filename}")

        except ET.ParseError as e:
            print(f"Error parsing {xml_filename}: {e}")
```

```python
from PIL import Image
import os

def resize_images(input_folder, output_folder, new_size=(260, 260)):
    # フォルダが存在しない場合は作成
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 入力フォルダ内の画像を処理
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)

        # 画像ファイルの場合のみ処理
        if os.path.isfile(input_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # 画像を開いてサイズを変更
            with Image.open(input_path) as img:
                resized_img = img.resize(new_size)

                # 出力フォルダに保存
                output_path = os.path.join(output_folder, filename)
                resized_img.save(output_path)

if __name__ == "__main__":
    # 入力と出力のフォルダを指定
    input_folder = "input_folder_path"
    output_folder = "output_folder_path"

    # 画像のサイズを指定
    new_size = (260, 260)

    # 画像のリサイズを実行
    resize_images(input_folder, output_folder, new_size)
```

```python
# 画像を260×260のpaddingしたものに変換
from PIL import Image
import os

def resize_with_padding(input_folder, output_folder, target_size=(260, 260)):
    # フォルダが存在しない場合は作成
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 入力フォルダ内の画像を処理
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)

        # 画像ファイルの場合のみ処理
        if os.path.isfile(input_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # 画像を開く
            with Image.open(input_path) as img:
                # 現在のサイズを取得
                original_width, original_height = img.size

                # ターゲットサイズに縦横比を維持したまま拡大・縮小
                if original_width > original_height:
                    new_width = target_size[0]
                    new_height = int(target_size[0] / original_width * original_height)
                else:
                    new_height = target_size[1]
                    new_width = int(target_size[1] / original_height * original_width)

                resized_img = img.resize((new_width, new_height), Image.ANTIALIAS)

                # パディングのサイズを計算
                width_diff = target_size[0] - new_width
                height_diff = target_size[1] - new_height

                left_padding = width_diff // 2
                top_padding = height_diff // 2
                right_padding = width_diff - left_padding
                bottom_padding = height_diff - top_padding

                # パディングを適用して新しい画像を作成
                padded_img = Image.new('RGB', target_size, (0, 0, 0))  # 黒い背景で初期化
                padded_img.paste(resized_img, (left_padding, top_padding))

                # 出力フォルダに保存
                output_path = os.path.join(output_folder, filename)
                padded_img.save(output_path)

if __name__ == "__main__":
    # 入力と出力のフォルダを指定
    input_folder = "/content/drive/MyDrive/Manga109_released_2021_12_30/face_images/Belmondo/"
    output_folder = "/content/drive/MyDrive/Manga109_released_2021_12_30/face_images_260/Belmondo_260/"

    # ターゲットサイズを指定
    target_size = (260, 260)

    # パディングで画像のリサイズを実行
    resize_with_padding(input_folder, output_folder, target_size)

```
